.1 What is a random variable in probability theory?

A random variable is a numerical outcome of a random experiment. It maps outcomes to real numbers.


2. What are the types of random variables?
- Discrete Random Variable
- Continuous Random Variable

3. What is the difference between discrete and continuous distributions?
- Discrete distributions deal with countable outcomes (e.g., number of heads in coin tosses).
- Continuous distributions deal with outcomes over a continuum (e.g., height, weight).

4. What are probability distribution functions (PDF)?

PDF describes the likelihood of a continuous random variable assuming a particular value.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

    PDF shows the probability density.

    CDF gives cumulative probability up to a value.

6. What is a discrete uniform distribution?

A distribution where all outcomes have equal probability (e.g., rolling a fair die).

7. What are the key properties of a Bernoulli distribution?

    Two outcomes: success (1) or failure (0)

    Mean = p, Variance = p(1 - p)

8. What is the binomial distribution, and how is it used in probability?

Used to model the number of successes in a fixed number of independent Bernoulli trials.

9. What is the Poisson distribution and where is it applied?

It models the number of occurrences of an event in a fixed interval (used in queuing theory, call centers, etc.).

10. What is a continuous uniform distribution?

Probability is evenly distributed over an interval [a, b].

11. What are the characteristics of a normal distribution?

    Bell-shaped curve

    Symmetrical about the mean

    Mean = Median = Mode

12. What is the standard normal distribution, and why is it important?

A normal distribution with mean 0 and SD 1. Used in Z-score calculations and standardization.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the mean of a large enough sample will be approximately normal, regardless of the population's distribution.

14. How does the Central Limit Theorem relate to the normal distribution?

It explains why normal distribution appears frequently even when population data isn't normal.

15. What is the application of Z statistics in hypothesis testing?

Used to test the significance of sample results under known variance and large sample sizes.

16. How do you calculate a Z-score, and what does it represent?

Z = (X - μ) / σ
It tells how many standard deviations a value is from the mean.

17. What are point estimates and interval estimates in statistics?

    Point estimate: A single value estimate (e.g., sample mean).

    Interval estimate: A range with a confidence level (e.g., 95% CI).

18. What is the significance of confidence intervals in statistical analysis?

They indicate the range in which a population parameter is likely to fall with a certain level of confidence.


19. What is the relationship between a Z-score and a confidence interval?

Z-scores help determine the bounds of confidence intervals in the standard normal distribution.


20. How are Z-scores used to compare different distributions?

They standardize data, allowing comparison across different units or scales.


21. What are the assumptions for applying the Central Limit Theorem?

    Independent observations

    Sufficiently large sample size (n ≥ 30)

    Identically distributed variables

22. What is the concept of expected value in a probability distribution?

The expected value is the theoretical average of a random variable, calculated as the sum of all possible values weighted by their probabilities.

The expected outcome (or expected value) is the weighted average of these possible values, where each value is multiplied by its probability.
```python
#3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x: The outcome (0 or 1)
    - p: The probability of success (0 <= p <= 1)

    Returns:
    - Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage:
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the discrete uniform distribution
a = 1  # Lower bound of the distribution
b = 10  # Upper bound of the distribution

# Generate the x values (discrete points from a to b)
x = np.arange(a, b + 1)

# Calculate the probabilities for each value (since it's a uniform distribution)
pmf = np.ones_like(x) / len(x)

# Calculate the CDF by cumulatively summing the PMF
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue')

# Labels and title
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.grid(True)
plt.xticks(x)  # Show all x values as ticks

# Show the plot
plt.legend()
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

import numpy as np
import matplotlib.pyplot as plt

# Function to demonstrate CLT using a non-normal distribution (Uniform distribution)
def clt_simulation(population_size, sample_size, num_samples):
    # Generate data from a uniform distribution
    population = np.random.uniform(0, 10, population_size)

    # List to store the means of each sample
    sample_means = []

    # Take samples and calculate their means
    for _ in range(num_samples):
        sample = np.random.choice(population, sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Convert sample means to a numpy array
    sample_means = np.array(sample_means)

    # Plot the histogram of the sample means (CLT)
    plt.figure(figsize=(8, 6))
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

    # Plot the theoretical normal distribution for comparison
    mean = np.mean(sample_means)
    std_dev = np.std(sample_means)
    x = np.linspace(min(sample_means), max(sample_means), 1000)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, pdf, label='Theoretical Normal PDF', color='red', lw=2)

    # Labels and title
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.title('Central Limit Theorem Demonstration (Uniform Distribution)')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Parameters for the simulation
population_size = 10000  # Size of the population
sample_size = 50         # Size of each sample
num_samples = 1000       # Number of samples to take

# Run the CLT simulation
clt_simulation(population_size, sample_size, num_samples)

```

```python
#18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate the Z-score for a given data point and compare it to a standard normal distribution
def z_score_comparison(data, data_point):
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score for the given data point
    z_score = (data_point - mean) / std_dev

    # Print the Z-score
    print(f"Z-score for the data point {data_point}: {z_score}")

    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)  # PDF of standard normal distribution (mean=0, std=1)

    # Plot the standard normal distribution and the Z-score line
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf, label='Standard Normal Distribution (mean=0, std=1)', color='b', lw=2)
    plt.axvline(x=z_score, color='r', linestyle='--', label=f'Z-score = {z_score:.2f}')

    # Labels and title
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.title('Z-score Comparison with Standard Normal Distribution')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Example data and a specific data point
data = np.random.normal(50, 10, 1000)  # Example data: Normally distributed data with mean=50 and std=10
data_point = 60  # The data point for which we want to calculate the Z-score

# Call the function to calculate and compare the Z-score
z_score_comparison(data, data_point)
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

# Function to perform Z-test for hypothesis testing
def z_test(sample, population_mean, population_std_dev, alpha=0.05):
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate critical Z value for two-tailed test at significance level alpha
    critical_value = stats.norm.ppf(1 - alpha/2)

    # Perform hypothesis testing
    print(f"Z-statistic: {z_stat}")
    print(f"Critical value for alpha={alpha}: ±{critical_value}")

    if abs(z_stat) > critical_value:
        print("Reject the null hypothesis (There is a significant difference).")
    else:
        print("Fail to reject the null hypothesis (No significant difference).")

    # Return Z-statistic and critical value for further analysis
    return z_stat, critical_value

# Sample dataset (e.g., sample data of test scores)
sample_data = np.random.normal(75, 10, 100)  # Normally distributed data with mean=75, std=10, and sample size=100

# Assumed population mean and standard deviation
population_mean = 70  # Assumed population mean (e.g., average test score)
population_std_dev = 10  # Assumed population standard deviation

# Perform Z-test
z_stat, critical_value = z_test(sample_data, population_mean, population_std_dev)
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate and interpret the confidence interval
def confidence_interval(data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (use ddof=1 for unbiased estimate)
    sample_size = len(data)

    # Calculate the critical value for the chosen confidence level
    alpha = 1 - confidence_level
    critical_value = stats.norm.ppf(1 - alpha / 2)  # Z-value for two-tailed test

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Standard Deviation: {sample_std}")
    print(f"Confidence Level: {confidence_level * 100}%")
    print(f"Margin of Error: {margin_of_error}")
    print(f"Confidence Interval: ({lower_bound}, {upper_bound})")

    # Plot the confidence interval
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=20, alpha=0.6, color='blue', edgecolor='black')
    plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
    plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Lower Bound of CI')
    plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper Bound of CI')

    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.title(f'Confidence Interval Visualization (Confidence Level: {confidence_level * 100}%)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example dataset (e.g., sample data of test scores)
data = np.random.normal(75, 10, 100)  # Normally distributed data with mean=75, std=10, and sample size=100

# Call the function to calculate and plot the confidence interval
confidence_interval(data, confidence_level=0.95)
```

```python
#21.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate and interpret the confidence interval for the sample mean
def confidence_interval_normal_data(sample_data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (use ddof=1 for unbiased estimate)
    sample_size = len(sample_data)

    # Calculate the critical value using the T-distribution for a sample (since we don't know the population std)
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)  # t-value for two-tailed test with df = sample_size - 1

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Standard Deviation: {sample_std}")
    print(f"Confidence Level: {confidence_level * 100}%")
    print(f"Margin of Error: {margin_of_error}")
    print(f"Confidence Interval: ({lower_bound}, {upper_bound})")

    # Plot the data and the confidence interval
    plt.figure(figsize=(8, 6))
    plt.hist(sample_data, bins=20, alpha=0.6, color='blue', edgecolor='black')
    plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
    plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Lower Bound of CI')
    plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper Bound of CI')

    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.title(f'Confidence Interval Visualization (Confidence Level: {confidence_level * 100}%)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate random data from a normal distribution (mean=50, std=10, n=100)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(50, 10, 100)  # 100 samples, mean=50, std=10

# Call the function to calculate and plot the confidence interval
confidence_interval_normal_data(sample_data, confidence_level=0.95)
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set the lambda (mean number of events)
lambda_val = 4  # average events per interval

# Generate a range of k values (discrete values)
k = np.arange(0, 15)

# Calculate CDF values
cdf_values = poisson.cdf(k, mu=lambda_val)

# Print CDF values
for i in range(len(k)):
    print(f"P(X ≤ {k[i]}) = {cdf_values[i]:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(k, cdf_values, where='post', label=f'Poisson CDF (λ = {lambda_val})', color='green')
plt.scatter(k, cdf_values, color='green')
plt.title('Cumulative Distribution Function of Poisson Distribution')
plt.xlabel('k (number of events)')
plt.ylabel('P(X ≤ k)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#25.  Write a Python program to compare the standard deviations of two datasets and visualize the difference


import numpy as np
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(42)  # For reproducibility
data1 = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, StdDev=10
data2 = np.random.normal(loc=50, scale=20, size=1000)  # Mean=50, StdDev=20

# Compute standard deviations
std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation
std_dev2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram
axes[0].hist(data1, bins=30, alpha=0.7, label="Dataset 1")
axes[0].hist(data2, bins=30, alpha=0.7, label="Dataset 2")
axes[0].set_title("Histogram Comparison")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].legend()

# Boxplot
axes[1].boxplot([data1, data2], labels=["Dataset 1", "Dataset 2"])
axes[1].set_title("Boxplot Comparison")

plt.tight_layout()
plt.show()
```

```python
#26.  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np

# Generate a sample dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, StdDev=15

# Calculate the range
data_range = np.max(data) - np.min(data)

# Calculate the interquartile range (IQR)
q1 = np.percentile(data, 25)  # First quartile (Q1)
q3 = np.percentile(data, 75)  # Third quartile (Q3)
iqr = q3 - q1

# Display results
print(f"Range of dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#27.  Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate a sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, StdDev=15

# Apply Z-score normalization
normalized_data = zscore(data)

# Visualization: Before & After Normalization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Data Histogram
axes[0].hist(data, bins=30, alpha=0.7, color='blue', label="Original Data")
axes[0].set_title("Original Dataset")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].legend()

# Normalized Data Histogram
axes[1].hist(normalized_data, bins=30, alpha=0.7, color='green', label="Z-score Normalized Data")
axes[1].set_title("Z-score Normalized Dataset")
axes[1].set_xlabel("Z-score")
axes[1].set_ylabel("Frequency")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
#28.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """Calculate skewness and kurtosis of a dataset."""
    skewness = skew(data)
    kurt = kurtosis(data)  # Excess kurtosis (subtracts 3 from actual kurtosis)

    return skewness, kurt

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, StdDev=15

# Compute skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

# Display results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurt:.4f}")
```
